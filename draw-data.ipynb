```python
from pylab import *
import csv

from matplotlib.colors import LinearSegmentedColormap
colors = ['A70226','FB9D59','CCCC66','94C7DF','313695']
colors = ['#'+cc for cc in colors]
cmapK = LinearSegmentedColormap.from_list('cmapK', colors = colors, N=200)


# When re-storing in csv, the float number may vary slightly, E.g. 0.00001
```

```python
def get_coldata(file_name,col = [0,1,8] , is_reshape = None):
    csv_reader = csv.reader(open(file_name))

    csv_row = [row for row in csv_reader]

    csv_row = np.array(csv_row,dtype=float)
    data = []
    for l in col:
        data_add = csv_row[:,l]
        if is_reshape != None:
            data_add = data_add.reshape(is_reshape)
        data.append(data_add)
    return data
```

```python
is_reshape = (41,41)
exp2 = get_coldata(file_name='data//Fig2-a.csv',col =range(4), is_reshape = is_reshape)
x,y,sigy,sigz = exp2

def plot_style():
    fontsize = 15
    fontsize2 = 20
    plt.xticks(np.arange(0,2.5,0.5),fontsize = fontsize)
    plt.yticks(np.arange(-4,5,1),fontsize=fontsize)
    plt.ylim(-4,4.1)
    plt.xlim(-0.05,2.05)
    plt.xlabel('$k/\pi$',fontsize = fontsize2)
    plt.ylabel('$\mu$',fontsize = fontsize2)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)

plt.quiver(x,y,sigy,sigz,sigz,color = 'white',cmap =cmapK,units='x',
                   scale_units = 'x',pivot='mid', width=0.007,headlength =5,headwidth =5,
                   scale=20)

k0 = np.linspace(-0.1,2.1)
plt.plot(k0,k0*0+2,'r--',linewidth=3)
plt.plot(k0,k0*0-2,'r--',linewidth=3)
plot_style()
```

```python
is_reshape = (41,41)
exp3s = []

for i in range(4):
    exp3 = get_coldata(file_name='data//Fig3-'+str('abcd'[i])+'.csv',col =range(4), is_reshape = is_reshape)
    exp3s += [exp3]

def plot_style():
    fontsize = 15
    fontsize2 = 20
    plt.xticks(np.arange(0,2.5,0.5),fontsize=fontsize)
    plt.yticks(np.arange(-4,6,2),fontsize=fontsize)
    plt.ylim(-5.1,4.1)
    plt.xlim(-0.025,2.025)
    plt.xlabel('$k/\pi$',fontsize = fontsize2)
    plt.ylabel('$\mu$',fontsize = fontsize2)

fig = plt.figure(figsize=(8,8))

lamb2 = [0.2,0.4,0.6,1.0]

for i in range(4):
    ax = fig.add_subplot(2,2,1+i)
    x,y,sigy,sigz = exp3s[i]
    plt.quiver(x,y,sigy,sigz,sigz,color = 'white',cmap =cmapK,units='x',
                       scale_units = 'x',pivot='mid', width=0.007,headlength =5,headwidth =5,
                       scale=20)
    
    k0 = np.linspace(-0.1,2.1)
    
    linewidth = 2.5
    color = 'crimson'
    plt.plot(k0,k0*0+2-2*lamb2[i],'--',color=color,linewidth=linewidth)
    plt.plot(k0,k0*0-2-2*lamb2[i],'--',color=color,linewidth=linewidth)

    if i >= 2:
        plt.plot(k0,k0*0+2*lamb2[i],'--',color=color,linewidth=linewidth)
    plot_style()

#     if i in [0,1]:
#         plt.xticks([])
#         plt.xlabel(None,color='w')
#     if i in [1,3]:
#         plt.yticks([])
#         plt.ylabel(None,color='w')

fig.tight_layout()
```

```python
is_reshape = (41,41)
exp4s = []

for i in range(3):
    exp4 = get_coldata(file_name='data//Fig4-'+str('abc'[i])+'.csv',col =range(4), is_reshape = is_reshape)
    exp4s += [exp4]

def plot_style():
    fontsize = 15
    fontsize2 = 20
    plt.xticks(np.arange(0,2.5,0.5),fontsize=fontsize)
    plt.yticks(np.arange(-4,6,2),fontsize=fontsize)
    plt.ylim(-4.1,4.1)
    plt.xlim(-0.025,2.025)
    plt.xlabel('$k/\pi$',fontsize = fontsize2)
    plt.ylabel('$\lambda_2$',fontsize = fontsize2)

fig = plt.figure(figsize=(10,4))

lamb1 = [0,1,2]

for i in range(3):
    ax = fig.add_subplot(1,3,1+i)
    x,y,sigy,sigz = exp4s[i]
    plt.quiver(x,y,sigy,sigz,sigz,color = 'white',cmap =cmapK,units='x',
                       scale_units = 'x',pivot='mid', width=0.007,headlength =5,headwidth =5,
                       scale=20)
    
    k0 = np.linspace(-0.1,2.1)
    
    linewidth = 2.5
    color = 'crimson'
    plt.plot(k0,k0*0-1,'--',color=color,linewidth=linewidth)
    plt.plot(k0,k0*0+1+lamb1[i],'--',color=color,linewidth=linewidth)
    plt.plot(k0,k0*0+1-lamb1[i],'--',color=color,linewidth=linewidth)
    
    plot_style()

    if i in [1,2]:
        plt.yticks([])
        plt.ylabel(None,color='w')

fig.tight_layout()
```

```python
is_reshape = (41,41)
exp7s = []

for i in range(12):
    exp7 = get_coldata(file_name='data//Fig7-'+str(i+1)+'.csv',col =range(2), is_reshape = is_reshape)
    exp7s += [exp7]

def plot_style():
    fontsize = 12
    fontsize2 = 14
    plt.xticks(np.arange(0,2.5,0.5),fontsize=fontsize)
    plt.yticks(np.arange(-4,6,2),fontsize=fontsize)
    plt.ylim(-4.1,4.1)
    plt.xlim(-0.025,2.025)


fig = plt.figure(figsize=(12,8))

lamb1 = np.arange(0,2.4,0.2)

for i in range(12):
    ax = fig.add_subplot(3,4,1+i)
    sigy,sigz = exp7s[i]
    
    x = np.linspace(0,2,41) # k/pi
    y = np.linspace(-4,4,41) # lambda_2
    plt.quiver(x,y,sigy,sigz,sigz,color = 'white',cmap =cmapK,units='x',
                       scale_units = 'x',pivot='mid', width=0.007,headlength =5,headwidth =5,
                       scale=20)
    
    k0 = np.linspace(-0.1,2.1)
    
    linewidth = 2.5
    alpha = 1
    color = 'crimson'
    plt.plot(k0,k0*0-1,'--',color=color,linewidth=linewidth,alpha = alpha)
    plt.plot(k0,k0*0+1+lamb1[i],'--',color=color,linewidth=linewidth,alpha = alpha)
    plt.plot(k0,k0*0+1-lamb1[i],'--',color=color,linewidth=linewidth,alpha = alpha)
    
    plot_style()
    
    fontsize2 = 14
    if i%4 != 0:
        plt.yticks([])
    else:
        plt.ylabel('$\lambda_2$',fontsize = fontsize2)
        
    
    if i <= 7:
        plt.xticks([])
    else:
        plt.xlabel('$k/\pi$',fontsize = fontsize2)
    
    
fig.tight_layout()

```
